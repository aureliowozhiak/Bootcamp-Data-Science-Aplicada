# 01.1 - Aquisição e leitura de dados reais
## Analisando dados da saúde com Python e Pandas

Dataset: http://tabnet.datasus.gov.br/cgi/deftohtm.exe?sih/cnv/qiuf.def
Filtros aplicados:
 - Linha: Unidade da Federação
 - Coluna: Ano/Mês processamento
 - Conteúdo: Valor total
 - Períodos disponíveis: Jan/2008 até Mar/2021
 
 [<- Voltar para o repositório](https://github.com/aureliowozhiak/Bootcamp-Data-Science-Aplicada)

```python
#importando biblioteca pandas
import pandas as pd
```

```python
#lendo o arquivo csv com algumas especificações de enconding, padrão do separador ";",
#e identificador de milhares e decimal
#também foi usado os parametros skip pra pular as 3 primeiras e 12 ultimas linhas q não fazem parte dos dados em si!
dados = pd.read_csv("../datasets/A122751189_28_143_208.csv", encoding="ISO-8859-1",
                   skiprows = 3, sep=";", skipfooter=12,
                   thousands=".", decimal=",")
#mostrando apenas os primeiros dados do dataframe
dados.head()
```

```python
#mostrando os ultimos dados do dataframe
dados.tail()
```

```python
#informações do dataframe
dados.info()
```

```python
#formatando as casas decimais
pd.options.display.float_format = "{:.2f}".format
```

```python
#média de cada mês
dados.mean()
```

```python
#Filtro por mês específico
dados["2008/Jan"]
```

```python
#média filtrada por mês específico
dados["2008/Jan"].mean()
```

### Plotando os dados em um gráfico

```python
#Podemos filtrar por um mês em específico
dados.plot(x="Unidade da Federação", y="2008/Ago")
```

```python
#Porém a melhor forma de mostrar esses dados não é usar um gráfico de linha, mas sim um gráfico de barra
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar")
```

```python
#O gráfico está muito pequeno, então podemos mudar o tamanho dele
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar",figsize=(10,6))
```

```python
#O pandas não é uma biblioteca gráfica, por conta disso, vamos importar uma biblioteca especifica para plotagem de gráficos
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar",figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
plt.title("Valores gastos pelo SUS em R$ por Unidade da Federação em Agosto de 2008")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
ax = dados.sort_values(by="2021/Mar", ascending=True).plot(x="Unidade da Federação", y="2021/Mar", kind="bar",figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
plt.title("Valores gastos pelo SUS em R$ por Unidade da Federação em Março de 2021")
plt.show()
```
